 Characters to be removed from certain columns to convert them to numeric values

```python
characters_remove = [',','₹','%']
columns = ['discounted_price','actual_price','discount_percentage','rating_count'] # Columns to be converted to float64 data type after removing characters
for ch in characters_remove: # Remove characters and convert selected columns to float64 data type
    for cl in columns:
        df[cl] = df[cl].str.replace(ch,'')

# Change data type of selected columns to float64
for cl in columns:
    df[cl] = df[cl].astype('float64')
```

```python
df[columns].dtypes
```

```python
df[df['rating'].str.contains('|',regex=False)]
```

 Some rows in the 'rating' column contain the character '|' which prevents conversion to float64  Replace '|' with '4' and convert the 'rating' column to float64 data type

```python
df['rating'] = df['rating'].str.replace('|','4').astype('float64')
```

 Check the data type of the 'rating' column after the conversion

```python
df['rating'].dtype
```

 Count the number of products in each category

```python
df['category'].value_counts()
```

 Preprocess the 'category' column to remove extra information and symbols

```python
df['category'] = df.category.apply(lambda x:x.split('|')[0].replace('&',' & '))
df['category'].value_counts()
```

 Visualize the count of products in each category using a countplot

```python
sns.countplot(data = df, y = 'category')
```

 Create a strip plot to show the demand for products in different price ranges for each category

```python
fig, ax = plt.subplots(figsize=(15, 5))
sns.set(style = 'whitegrid')
ax.set_title('Products Demand in Price Range', fontweight='bold')
sns.stripplot(ax =ax, data = df, y = 'category', x= 'actual_price',hue="category", jitter =0.3,legend=False)
```

 Create a box plot to show the distribution of product ratings in different categories

```python
fig, ax = plt.subplots(figsize=(15, 5))
sns.set(style = 'whitegrid')
ax.set_title('Ratings of Products In Categories', fontweight='bold')
sns.boxplot(ax =ax, data = df, y = 'category', x= 'rating')
```

Create a bar plot to display the top discount percentage categories

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.set(style = 'whitegrid')
discount_percentage = df.groupby('category')['discount_percentage'].apply(lambda x: x.mode()).sort_values(ascending = False).reset_index()
ax.set_title('Top Discount Percentage Categories',fontweight='bold' )
sns.barplot(ax=ax,data = discount_percentage, y =discount_percentage.category,  x = discount_percentage.discount_percentage,palette='Blues_r')
ax.bar_label(ax.containers[0],fmt='%.0f%%')
```

 Calculate the sales for each product by multiplying discounted_price with rating_count

```python
df['sales'] =  df.discounted_price*df.rating_count
df.sales.head()
```

 Calculate the total sales for each category and sort in descending order

```python
total_sales  = df.groupby('category')['sales'].sum().sort_values(ascending = False)
```

Function to convert large numbers to human-readable format

```python
import math

millnames = ['',' Thousand',' Million',' Billion',' Trillion']

def millify(n):
    n = float(n)
    millidx = max(0,min(len(millnames)-1,
                        int(math.floor(0 if n == 0 else math.log10(abs(n))/3))))

    return '{:.0f}{}'.format(n / 10**(3 * millidx), millnames[millidx])
```

 Convert total sales to human-readable format

```python
values = []
for x in total_sales.values:
    values.append(millify(x))
    print(millify(x))
```

 Create a bar plot to show the total sales of products in different categories

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.set_title('Total Sales of Categories',fontweight='bold' )
sns.barplot(ax= ax,y =total_sales.index,x =total_sales.values,palette='Blues_r')
ax.set_ylabel(' Categories', fontsize = 20)
ax.bar_label(ax.containers[0], labels=values, padding=3)
```
